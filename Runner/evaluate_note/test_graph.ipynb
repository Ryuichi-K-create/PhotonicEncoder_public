```python
import matplotlib.pyplot as plt

titlefont = 23
fontsize = 23
labelfont = 20
# データ
models = ["(1)Simple-Encoder \n + 2-Layer MLP",
          "(2)DEQ-Encoder \n + 1-Layer Classifier",
          "(3)DEQ-Encoder \n + 2-Layer MLP"]
params = [3850, 2950, 6300]
acc = [76.33, 77.40, 79.49]

# カラースタイル
colors = ['#1f4e79',  # 濃い青 (Deep Blue)
           '#c45a00',  # 濃いオレンジ (Dark Orange)
           '#1b5e20']  # 濃い緑 (Deep Green)
markers = ['o', 's', '^']

# 図の作成
plt.figure(figsize=(7,5))
for i in range(len(models)):
    plt.scatter(params[i], acc[i], 
                color=colors[i], s=150, marker=markers[i],
                edgecolor='black', linewidth=1.2, zorder=3)

# 軽いトレンド線（破線）
plt.plot(params, acc, color='gray', linestyle='--', linewidth=1)

# 軸ラベル・タイトル
plt.title("Parameter–Accuracy Trade-off Improved by DEQ (1:16)", fontsize=titlefont, pad=12)
plt.xlabel("Number of Parameters", fontsize=fontsize)
plt.ylabel("Accuracy [%]", fontsize=fontsize)

# 軸範囲（精度差を強調）
plt.ylim(75.5, 80)
plt.xlim(2500, 7000)

# グリッド
plt.grid(True, linestyle=':', alpha=0.6)

# 各点にラベル付け
for i, label in enumerate(models):
    plt.text(params[i]+80, acc[i]-0.1, label,
             fontsize=labelfont, va='center', color=colors[i])

# # 矢印説明（Optional）
# plt.annotate("Higher accuracy\nwith fewer parameters",
#              xy=(2950, 77.4), xytext=(4200, 78.8),
#              arrowprops=dict(arrowstyle="->", color='black'),
#              fontsize=10, ha='left')

plt.tight_layout()
plt.show()
```
