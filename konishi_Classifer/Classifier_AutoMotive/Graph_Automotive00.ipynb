```python
import csv
import os
import matplotlib.pyplot as plt
import numpy as np
import ast
home_directory = os.path.expanduser('~')
onedrive_folder_name = 'OneDrive'
onedrive_path = os.path.join(home_directory, onedrive_folder_name)
take_directory = os.path.join(onedrive_path,'CODES', 'konishi_AutoEncoder', 'AutoMotive_data_001','texts') 

from matplotlib import rcParams
rcParams['font.family'] = 'Times New Roman'

colors = [
    "#1f77b4",  # 青
    "#ff7f0e",  # オレンジ
    "#2ca02c",  # 緑
    "#d62728",  # 赤
    "#9467bd",  # 紫
    "#8c564b",  # 茶
    "#e377c2",  # ピンク
    "#7f7f7f",  # グレー
    "#bcbd22",  # 黄緑
    "#17becf",  # 水色
]
```

```python
# CSVファイルのパス
file_name = 'MLP_VCR_PM_Class00_list.csv'
full_path = os.path.join(take_directory, file_name)

# CSVファイルからデータを読み込む
with open(full_path, 'r') as file:
    reader = csv.reader(file)
    rows = list(reader)

# 各行を評価してリスト形式に変換
All_last_LOSSs_ = np.array([eval(row) for row in rows[0]])  # 1行目がMSEデータ
All_last_ACCs_ = np.array([eval(row) for row in rows[1]])  # 2行目がSSIMデータ

# NumPy配列に変換
All_last_LOSSs_ = np.array(All_last_LOSSs_)
All_last_ACCs_ = np.array(All_last_ACCs_)

# データを2次元配列に変換（必要に応じて）
LOSS_means = np.mean(All_last_LOSSs_, axis=1)  # 各 leverage に対する平均
LOSS_stds = np.std(All_last_LOSSs_, axis=1)   # 各 leverage に対する標準偏差
ACC_means = np.mean(All_last_ACCs_, axis=1)  # 各 leverage に対する平均
ACC_stds = np.std(All_last_ACCs_, axis=1)   # 各 leverage に対する標準偏差

#-------------------------------------------------------------------------------------
# CSVファイルのパス
file_name = 'MLP_VCR_IM_Class00_list.csv'
full_path = os.path.join(take_directory, file_name)

with open(full_path, 'r') as file:
    reader = csv.reader(file)
    rows = list(reader)

# 各行を評価してリスト形式に変換
All_last_LOSSs_ = np.array([eval(row) for row in rows[0]])  # 1行目がMSEデータ
All_last_ACCs_ = np.array([eval(row) for row in rows[1]])  # 2行目がSSIMデータ

# NumPy配列に変換
All_last_LOSSs_ = np.array(All_last_LOSSs_)
All_last_ACCs_ = np.array(All_last_ACCs_)

# データを2次元配列に変換（必要に応じて）
LOSS_means2 = np.mean(All_last_LOSSs_, axis=1)  # 各 leverage に対する平均
LOSS_stds2 = np.std(All_last_LOSSs_, axis=1)   # 各 leverage に対する標準偏差
ACC_means2 = np.mean(All_last_ACCs_, axis=1)  # 各 leverage に対する平均
ACC_stds2 = np.std(All_last_ACCs_, axis=1)   # 各 leverage に対する標準偏差

#-------------------------------------------------------------------------------------
# CSVファイルのパス
file_name = 'MLP_VCR_MZM_Class00_list.csv'
full_path = os.path.join(take_directory, file_name)

with open(full_path, 'r') as file:
    reader = csv.reader(file)
    rows = list(reader)

# 各行を評価してリスト形式に変換
All_last_LOSSs_ = np.array([eval(row) for row in rows[0]])  # 1行目がMSEデータ
All_last_ACCs_ = np.array([eval(row) for row in rows[1]])  # 2行目がSSIMデータ

# NumPy配列に変換
All_last_LOSSs_ = np.array(All_last_LOSSs_)
All_last_ACCs_ = np.array(All_last_ACCs_)

# データを2次元配列に変換（必要に応じて）
LOSS_means3 = np.mean(All_last_LOSSs_, axis=1)  # 各 leverage に対する平均
LOSS_stds3 = np.std(All_last_LOSSs_, axis=1)   # 各 leverage に対する標準偏差
ACC_means3 = np.mean(All_last_ACCs_, axis=1)  # 各 leverage に対する平均
ACC_stds3 = np.std(All_last_ACCs_, axis=1)   # 各 leverage に対する標準偏差

#-------------------------------------------------------------------------------------
# CSVファイルのパス
file_name = 'MLP_VCR_LI_Class00_list.csv'
full_path = os.path.join(take_directory, file_name)

with open(full_path, 'r') as file:
    reader = csv.reader(file)
    rows = list(reader)

# 各行を評価してリスト形式に変換
All_last_LOSSs_ = np.array([eval(row) for row in rows[0]])  # 1行目がMSEデータ
All_last_ACCs_ = np.array([eval(row) for row in rows[1]])  # 2行目がSSIMデータ

# NumPy配列に変換
All_last_LOSSs_ = np.array(All_last_LOSSs_)
All_last_ACCs_ = np.array(All_last_ACCs_)

# データを2次元配列に変換（必要に応じて）
LOSS_means4 = np.mean(All_last_LOSSs_, axis=1)  # 各 leverage に対する平均
LOSS_stds4 = np.std(All_last_LOSSs_, axis=1)   # 各 leverage に対する標準偏差
ACC_means4 = np.mean(All_last_ACCs_, axis=1)  # 各 leverage に対する平均
ACC_stds4 = np.std(All_last_ACCs_, axis=1)   # 各 leverage に対する標準偏差

#-------------------------------------------------------------------------------------
# leverageの範囲 (1:2, 1:4, ..., 1:16)
leverages = [1,2,3,6]
memory_lis =leverages

# LOSSのグラフ
fig1, ax1 = plt.subplots()
ax1.tick_params(axis='both', labelsize=15)
ax1.errorbar(
    x=leverages, y=LOSS_means, yerr=LOSS_stds,
    fmt='-o', color=colors[0], ecolor=colors[0], capsize=5, label='(1)Phase Modulation'
)
ax1.errorbar(
    x=leverages, y=LOSS_means2, yerr=LOSS_stds2,
    fmt='-^', color=colors[1], ecolor=colors[1], capsize=5, label='(2)Intensity Modulation'
)

ax1.errorbar(
    x=leverages, y=LOSS_means3, yerr=LOSS_stds3,
    fmt='-^', color=colors[2], ecolor=colors[2], capsize=5, label='(3)MZ Modulation'
)
ax1.errorbar(
    x=leverages, y=LOSS_means4, yerr=LOSS_stds4,
    fmt='-^', color=colors[3], ecolor=colors[3], capsize=5, label='(4)Linear'
)

ax1.set_xlabel('Compression Ratio ', fontsize=15)
ax1.set_xticks(memory_lis)
ax1.set_xticklabels([f"1:{x}" for x in memory_lis])
ax1.set_ylabel('LOSS' ,fontsize=15)
#ax1.set_title('MSE')
ax1.legend(fontsize=15, loc='upper left', bbox_to_anchor=(1.0, 1))
ax1.grid(True)
plt.show()

# SSIMのグラフ
fig2, ax2 = plt.subplots()
ax2.tick_params(axis='both', labelsize=15)
ax2.errorbar(
    x=leverages, y=ACC_means, yerr=ACC_stds,
    fmt='-o', color=colors[0], ecolor=colors[0], capsize=5, label='(1)Phase Modulation'
)
ax2.errorbar(
    x=leverages, y=ACC_means2, yerr=ACC_stds2,
    fmt='-^', color=colors[1], ecolor=colors[1], capsize=5, label='(2)Intensity Modulation'
)

ax2.errorbar(
    x=leverages, y=ACC_means3, yerr=ACC_stds3,
    fmt='-^', color=colors[2], ecolor=colors[2], capsize=5, label='(3)MZ Modulation'
)

ax2.errorbar(
    x=leverages, y=ACC_means4, yerr=ACC_stds4,
    fmt='-^', color=colors[3], ecolor=colors[3], capsize=5, label='(4)Linear'
)

ax2.set_xlabel('Compression Ratio ', fontsize=15)
ax2.set_xticks(memory_lis)
ax2.set_xticklabels([f"1:{x}" for x in memory_lis])
ax2.set_ylabel('Accuracy', fontsize=15)
ax2.legend(fontsize=15, loc='upper left', bbox_to_anchor=(1.0, 1))
ax2.grid(True)
plt.show()
```
