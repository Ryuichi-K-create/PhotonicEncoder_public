```python
import numpy as np
import matplotlib.pyplot as plt

#  (1:N)
leverage_ratios = [50, 100, 150, 200, 250]

#  (1/leverage)
x_values = [1 / leverage for leverage in leverage_ratios]

#  ()
mse_means = [30, 25, 22, 20, 19]  # MSE
mse_stds = [5, 3, 2, 2, 1.5]      # MSE

ssim_means = [0.9, 0.85, 0.8, 0.78, 0.75]  # SSIM
ssim_stds = [0.05, 0.04, 0.03, 0.03, 0.02]  # SSIM

# 
fig, ax = plt.subplots(figsize=(8, 6))

# MSE
ax.errorbar(
    x=x_values, y=mse_means, yerr=mse_stds,
    fmt='-o', color='red', ecolor='red', capsize=5, label='MSE'
)

# SSIM
ax.errorbar(
    x=x_values, y=ssim_means, yerr=ssim_stds,
    fmt='-o', color='blue', ecolor='blue', capsize=5, label='SSIM'
)

# 
ax.set_xlabel('Compression Ratio (1:leverage)', fontsize=12)
ax.set_xticks(x_values)
ax.set_xticklabels(["1:%d" % leverage for leverage in leverage_ratios], fontsize=10)  # 1:N

# 
ax.set_ylabel('Metric Value', fontsize=12)

# 
ax.grid(visible=True, linestyle='--', alpha=0.6)

# 
plt.title('MSE and SSIM with Compression Ratio', fontsize=14)
plt.legend(fontsize=10)

# 
file_name = 'Metric_vs_Compression_Ratio.png'
plt.savefig(file_name, dpi=100, bbox_inches='tight')

# 
plt.show()
```
