```python
import csv
import os
import matplotlib.pyplot as plt
import numpy as np

home_directory = os.path.expanduser('~')
onedrive_folder_name = 'OneDrive'
onedrive_path = os.path.join(home_directory, onedrive_folder_name)
take_directory = os.path.join(onedrive_path,'CODES', 'konishi_AutoEncoder', 'AutoEncoder_data','texts') 
```

```python
#10/08

file_name1 = 'AE[5_]_list.csv'##
full_path1 = os.path.join(take_directory, file_name1)

# CSVファイルからデータを読み込む
list1_1 = []
list1_2 = []

with open(full_path1, mode='r') as file:
    reader = csv.reader(file)
    # 各行をリストに格納
    list1_1 = list(map(float, next(reader)))
    list1_2 = list(map(float, next(reader)))
    list1_1 = list1_1[::-1]
    list1_2 = list1_2[::-1]


file_name2 = 'AE[3_]_list.csv'##
full_path2 = os.path.join(take_directory, file_name2)

# CSVファイルからデータを読み込む
list2_1 = []
list2_2 = []

with open(full_path2, mode='r') as file:
    reader = csv.reader(file)
    # 各行をリストに格納
    list2_1 = list(map(float, next(reader)))
    list2_2 = list(map(float, next(reader)))
    list2_1 = list2_1[::-1]
    list2_2 = list2_2[::-1]

# グラフを描画

x_values = [x / 784 for x in range(len(list1_1))] 

plt.plot(x_values, list1_1, label='Complex Encoder:Train',color = 'cyan')
plt.plot(x_values, list1_2, label='Complex Encoder:Test',color = 'blue')

plt.plot(x_values, list2_1, label='Real Encoder:Train',color = 'red')
plt.plot(x_values, list2_2, label='Real Encoder:Test',color = 'orange')
plt.xlabel('N/M')
plt.ylabel('MSE')
plt.xlim(0,1)
plt.title('MSE Transition')
plt.legend()
plt.show()
```

```python
#11/27

file_name1 = 'AE[7-1]_repList.csv'##
full_path1 = os.path.join(take_directory, file_name1)

# CSVファイルからデータを読み込む
list1_1 = []
list1_2 = []

with open(full_path1, mode='r') as file:
    reader = csv.reader(file)
    # 各行をリストに格納
    list1_1 = list(map(float, next(reader)))
    list1_2 = list(map(float, next(reader)))
    list1_1 = list1_1[::-1]
    list1_2 = list1_2[::-1]


file_name2 = 'AE[7-1]_rep3List.csv'##
full_path2 = os.path.join(take_directory, file_name2)

# CSVファイルからデータを読み込む
list2_1 = []
list2_2 = []

with open(full_path2, mode='r') as file:
    reader = csv.reader(file)
    # 各行をリストに格納
    list2_1 = list(map(float, next(reader)))
    list2_2 = list(map(float, next(reader)))
    list2_1 = list2_1[::-1]
    list2_2 = list2_2[::-1]


file_name3 = 'AE[7-1$]_repList.csv'##
full_path3 = os.path.join(take_directory, file_name3)

# CSVファイルからデータを読み込む
list3_1 = []
list3_2 = []

with open(full_path3, mode='r') as file:
    reader = csv.reader(file)
    # 各行をリストに格納
    list3_1 = list(map(float, next(reader)))
    list3_2 = list(map(float, next(reader)))
    list3_1 = list3_1[::-1]
    list3_2 = list3_2[::-1]

# グラフを描画
DofIN = 784
width = 8

desired_ratios = ["1:1", "1:2", "1:4", "1:8","1:16"]
desired_N_values = [DofIN // int(ratio.split(":")[1]) for ratio in desired_ratios] 
x_values = np.arange(DofIN, 0, -width)[::-1]

#plt.plot(x_values, list1_1, label='Complex Encoder:Train',color = 'cyan')
plt.plot(x_values, list3_2, label='1-layer Encoder',color = 'green')

plt.plot(x_values, list1_2, label='2-layer Encoder',color = 'blue')

plt.plot(x_values, list2_2, label='3-layer Encoder',color = 'red')
plt.xlabel('N:M')
plt.ylabel('MSE')
plt.xticks(ticks=desired_N_values, labels=desired_ratios)
plt.title('MSE Transition')
plt.legend()
plt.show()


plt.plot(x_values, list3_1, label='1-layer Encoder',color = 'green')

plt.plot(x_values, list1_1, label='2-layer Encoder',color = 'blue')

plt.plot(x_values, list2_1, label='3-layer Encoder',color = 'red')
plt.xlabel('N:M')
plt.ylabel('SSIM')
plt.xticks(ticks=desired_N_values, labels=desired_ratios)
plt.title('SSIm Transition')
plt.legend()
plt.show()
```
