```python
import csv
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

home_directory = os.path.expanduser('~')
onedrive_folder_name = 'OneDrive'
onedrive_path = os.path.join(home_directory, onedrive_folder_name)
take_directory = os.path.join(onedrive_path,'Codes', 'konishi_AutoEncoder', 'AutoEncoder_data004ex','texts') 
```

```python
x_values = range(1,11)

custom_colors = [
    'red', 'blue', 'green', 'orange', 'purple',
    'cyan', 'magenta', 'yellow', 'brown', 'gray'
]

csv_files = glob.glob(os.path.join(take_directory, "*_2.csv"))

print(csv_files[0])

for file_idx, file_name1 in enumerate(csv_files):
    #full_path1 = os.path.join(take_directory, file_name1)
    file_name = os.path.basename(file_name1)  # 
    label_name = file_name.replace("_2.csv", "")#.replace("Linear", "")
    list1 = []
    with open(file_name1, mode='r') as file:
        reader = csv.reader(file)
        list1 = list(map(float, next(reader)))
        plt.plot(x_values, list1, label=label_name,color = custom_colors[file_idx])

plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.xlim(0,11)
plt.title('MSE Transition(1:8)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1)) 
plt.show()
```
